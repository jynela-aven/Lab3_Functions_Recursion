```python
# [CELL 1]
def compute_access_level(control_num, favorite_artist_length):
    MODULUS = 97
    return (control_num * favorite_artist_length) % MODULUS

def validate_access(access_level, control_num):
    threshold = control_num + 5
    return access_level >= threshold

def audit_log(func):
    def wrapper(*args, **kwargs):
        print("Authorization Started")
        result = func(*args, **kwargs)
        print("Authorization Completed")
        return result
    return wrapper

CONTROL_NUM = 14
FAVORITE_ARTIST = "RUEL"

FAVORITE_ARTIST_LENGTH = len(FAVORITE_ARTIST)

print(FAVORITE_ARTIST_LENGTH)

@audit_log
def authorize(control_num, artist_length):
    level = compute_access_level(control_num, artist_length)
    print(f"Computed Access Level: {level}")
    
    decision = validate_access(level, control_num)

    return "ACCESS GRANTED" if decision else "ACCESS DENIED"

final_decision = authorize(CONTROL_NUM, FAVORITE_ARTIST_LENGTH)
print(f"Final Authorization Decision: {final_decision}")

```
